```python
# Type Casting
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Workout_Type'] = df['Workout_Type'].astype('category')
print(df.dtypes)
```

```python
# Create the count plot
plt.figure(figsize=(8, 5))
sns.countplot(x='Workout_Type', data=df, palette='Set2')

# Add titles and labels
plt.title('Count of Each Workout Type')
plt.xlabel('Workout Type')
plt.ylabel('Count')

# Display the plot
plt.tight_layout()
plt.show()
```

Majority of users prefer a specific workout type, indicating their main fitness focus.










```python
# Create a histogram for the 'Steps' column
plt.figure(figsize=(8, 5))
sns.histplot(df['Steps'], bins=10, kde=True, color='pink')

# Add titles and labels
plt.title('Distribution of Steps')
plt.xlabel('Steps')
plt.ylabel('Frequency')

# Display the plot
plt.tight_layout()
plt.show()
```

INSIGHTS FROM ABOVE VISUALIZATION

Most users take a moderate number of steps, likely in the mid-range.

The distribution shows a peak in common activity levels, with fewer users at very high or low extremes.

The shape indicates some variation in activity, possibly with mild skewness.

A few users may be outliers, taking significantly more or fewer steps.



```python
# Create the line chart
plt.figure(figsize=(10, 5))
sns.lineplot(x='Date', y='Steps', data=df, marker='o', color='teal')

# Add titles and labels
plt.title(f'Steps Over Time for User ID {df}')
plt.xlabel('Date')
plt.ylabel('Steps')

# Show the plot
plt.tight_layout()
plt.show()
```

INSIGHTS FROM ABOVE VISUALIZATION

Insight: If the line fluctuates a lot, the user lacks consistency.

Plots daily step trends over time.
Shows how physical activity changes day by day.

Trend Detection:- Clear view of increases or drops in step activity.
Spikes:- High points reveal days of extra activity or workouts
Drops:- Low points suggest rest or inactive days.
Consistency:- Smooth line = steady routine; sharp changes = irregular activity.
Progress Tracking:- Helpful to see if step count improves over time.





```python
# Bar Chart
plt.figure(figsize=(8,4))
plt.bar(df['Date'], df['Steps'], color='lightgreen')
plt.xlabel('Date')
plt.ylabel('Steps')
plt.title('Daily Step Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Bar Chart Visualization :- how step count varies day to day

 If users mostly choose cardio, other types may need better promotion or personalization.

Displays the number of steps taken each day.
Helps identify daily activity trends.

Activity Trend: Shows how step count varies day to day.
Most Active Days: Peaks highlight the most active dates.
Least Active Days: Small bars show days with low movement.
Inconsistency: Fluctuating bar heights suggest irregular activity.
Behavior Patterns: May reveal active weekends or lazy weekdays.

```python
# Create a boxplot for the 'Steps' column
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Steps'], color='skyblue')

# Add labels and title
plt.xlabel("Steps")
plt.title("Boxplot of Steps")
plt.tight_layout()
plt.show()
```

INSIGHTS FROM ABOVE VISUALIZATION

 Insight: Outliers can indicate days of over-exertion or inactivity.

The boxplot displays the distribution of daily steps.

It includes the median, interquartile range (IQR), and any outliers.


Median:- shows the typical daily step count.

IQR:- indicates most steps fall in a consistent range.

Outliers:- show unusually high or low activity days.

Skewness:- tells if step counts lean low or high.

Consistency:- is seen if the box is tight with few outliers.

```python
# HEAT MAP
# Encode categorical columns (if any)
df_encoded = pd.get_dummies(df, drop_first=True)

# Compute the correlation matrix
correlation_matrix = df_encoded.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Matrix (after encoding)")
plt.tight_layout()
plt.show()
```

More steps = more calories burned.

More steps = more distance covered.

More distance = more calories burned.

Date or time doesn’t affect steps or calories.

Some features are not related to each other.
